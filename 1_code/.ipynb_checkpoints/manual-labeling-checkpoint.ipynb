```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import math
from sklearn import metrics
import seaborn as sns
import itertools
import statistics
pd.set_option("display.max_colwidth", 500)

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# helper functions
def plot_confusion_matrix(y_true, y_pred, normalize=False):
    """
    Input:
        y_true    : array of true binary labels. Eg: [0, 1, 0, 0, 1].
        y_pred    : array of predicted probabilities. Eg: [0.05, 0.55, 0.2, 0.8, 0.95].
        normalize : whether to show counts of proportion in a cell
    
    Output:
        confusion matrix
    """
    cm = metrics.confusion_matrix(y_true, y_pred)
    
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    #plt.colorbar()
    classes = np.unique(y_true)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if(normalize):
        precision=4
        cm = np.round(cm.astype('float') / cm.sum(axis=1)[:, np.newaxis], precision)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    pass

fake_pattern = re.compile(".+_fake\.jpg$")
true_pattern = re.compile(".+_true\.jpg$")
def get_y_true(name):
    if fake_pattern.search(name):
        return 1
    elif true_pattern.search(name):
        return 0
    else:
        return -1
    pass

def y_to_int(label):
    if(label == 'fake'):
        return 1
    elif label == 'real':
        return 0
    else:
        return -1
    pass
```

```python
# read data
soumya_labels = '../00_source_data/human-labelers/soumya.csv'
preeti_labels = '../00_source_data/human-labelers/preeti.csv'
jashandeep_labels = '../00_source_data/human-labelers/jashandeep.csv'
forum_labels = '../00_source_data/human-labelers/forum.csv'
prabhjot_labels = '../00_source_data/human-labelers/prabhjot.csv'

soumya = pd.read_csv(soumya_labels); soumya.shape
preeti = pd.read_csv(preeti_labels); preeti.shape
jashandeep = pd.read_csv(jashandeep_labels); jashandeep.shape
forum = pd.read_csv(forum_labels); forum.shape
prabhjot = pd.read_csv(prabhjot_labels); prabhjot.shape
```

```python
# filter data
soumya = soumya[["Label", "External ID", "Seconds to Label"]]
soumya = soumya[soumya.Label !="{}"]
soumya = soumya.sort_values(by=["External ID"])
print(soumya.shape)

preeti = preeti[["Label", "External ID", "Seconds to Label"]]
preeti = preeti[preeti.Label !="{}"]
preeti = preeti.sort_values(by=["External ID"])
print(preeti.shape)

jashandeep = jashandeep[["Label", "External ID", "Seconds to Label"]]
jashandeep = jashandeep[jashandeep.Label !="{}"]
jashandeep = jashandeep.sort_values(by=["External ID"])
print(jashandeep.shape)

forum = forum[["Label", "External ID", "Seconds to Label"]]
forum = forum[forum.Label !="{}"]
forum = forum.sort_values(by=["External ID"])
print(forum.shape)

prabhjot = prabhjot[["Label", "External ID", "Seconds to Label"]]
prabhjot = prabhjot[prabhjot.Label !="{}"]
prabhjot = prabhjot.sort_values(by=["External ID"])
print(prabhjot.shape)
```

```python
# SOUMYA
# get true label
soumya_y_true = soumya["External ID"].apply(lambda name: get_y_true(name))
y_true = soumya_y_true

# get annotations
soumya_y_pred = soumya.Label.apply(lambda text: text[-9:-5])
soumya_y_pred = soumya_y_pred.apply(lambda label: y_to_int(label))

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plot_confusion_matrix(soumya_y_true, soumya_y_pred, normalize=True)
print('Soumya was {}% accurate.'.format(100*metrics.accuracy_score(soumya_y_true, soumya_y_pred)))

# AUC
soumya_fpr, soumya_tpr, thresholds = metrics.roc_curve(soumya_y_true, soumya_y_pred, pos_label=1)
soumya_auc = metrics.auc(soumya_fpr, soumya_tpr)
print('Soumya\'s AUC was {}.'.format(soumya_auc))
      
# time to plot
#plt.subplot(1, 2, 2)
#sns.boxplot(x=soumya['Seconds to Label'])
#soumya['Seconds to Label'].describe()
```

```python
mean_accuracy = statistics.mean([metrics.accuracy_score(soumya_y_true, soumya_y_pred), 
                                 metrics.accuracy_score(preeti_y_true, preeti_y_pred), 
                                 metrics.accuracy_score(jashandeep_y_true, jashandeep_y_pred), 
                                 metrics.accuracy_score(forum_y_true, forum_y_pred),
                                 metrics.accuracy_score(prabhjot_y_true, prabhjot_y_pred)])

print('Mean accuracy of all human labelers = {:2f}%.'.format(100 * mean_accuracy))
```

```python
std = statistics.stdev([metrics.accuracy_score(soumya_y_true, soumya_y_pred), 
                        metrics.accuracy_score(preeti_y_true, preeti_y_pred), 
                        metrics.accuracy_score(jashandeep_y_true, jashandeep_y_pred), 
                        metrics.accuracy_score(forum_y_true, forum_y_pred),
                        metrics.accuracy_score(prabhjot_y_true, prabhjot_y_pred)])

print('Standard deviation of individual accuracy scores is {}%.'.format(round(100 * std, 2)))
```

```python
print('Mean AUC of all human labelers = {}.'.format(statistics.mean([soumya_auc, preeti_auc, jashandeep_auc, forum_auc, prabhjot_auc])))
```

```python
# VOTING CLASSIFIER
y_preds = np.concatenate((soumya_y_pred.values.reshape((1, -1)), 
                          preeti_y_pred.values.reshape((1, -1)), 
                          jashandeep_y_pred.values.reshape((1, -1)), 
                          forum_y_pred.values.reshape((1, -1)),
                          prabhjot_y_pred.values.reshape((1, -1))), axis=0).T

y_pred = np.apply_along_axis(statistics.mode, 1, y_preds)

plt.figure(figsize=(5, 5))
plot_confusion_matrix(y_true, y_pred, normalize=True)
print('Human baseline was {}% accurate.'.format(100*metrics.accuracy_score(y_true, y_pred)))
```
